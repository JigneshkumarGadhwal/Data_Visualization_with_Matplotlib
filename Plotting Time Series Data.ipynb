## Changing the date format:

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn-v0_8-whitegrid')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates,y, linestyle='solid', linewidth=1)

plt.gcf().autofmt_xdate()  # get current figure

# changing the formating of date:
date_format = mpl_dates.DateFormatter('%b,%d,%Y')

plt.gca().xaxis.set_major_formatter(date_format) # get current axis

# data = pd.read_csv('data.csv')
# price_date = data['Date']
# price_close = data['Close']

# plt.title('Bitcoin Prices')
# plt.xlabel('Date')
# plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

## Using CSV file:

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn-v0_8-whitegrid')

# Data of bitcoin:
data = pd.read_csv(r"C:\Users\JERRY\Jupyter_Python_2K24_25\Matplotlib Python\Data\data_Time_series_data.csv")

data['Date'] = pd.to_datetime(data['Date']) # converting the Date column datatype from string to datetime
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid',linewidth=1)

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
