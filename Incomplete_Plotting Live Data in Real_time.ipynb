# Plotting Live Data:

```python
import random
import matplotlib.pyplot as plt
import pandas as pd
from itertools import count
```

```python
plt.style.use('fivethirtyeight')

x_vals = [0,1,2,3,4,5]
y_vals = [0,1,3,2,3,5]

plt.plot(x_vals,y_vals,linewidth=1)

plt.tight_layout()
plt.show()
```
