### Plotting all data as a bar chart:

We simply just can not use plt.bar for each data and try plotting them , because it is gonna give us the issue of overlapping plots of the data, which would not make any sense at all.

```python
# Using plt.bar for each data:  (Notice the problem)

plt.style.use('fivethirtyeight')

plt.bar(ages_x,dev_y,color='#444444', label = "All devs")  

plt.bar(ages_x,py_dev_y, color = "#008fd5", label ="Python")

plt.bar(ages_x,js_dev_y, color = '#e5ae38', label = "JavaScript")

plt.legend()

plt.xlabel("Ages")
plt.ylabel("Median salary in USD")
plt.title("Median salary in USD by Age")

plt.tight_layout()  # for the padding

plt.show()
```

**Solution:**

We can fulfil this objective by offsetting the "X" values each time we plot data....

```python
import numpy as np
```

```python
plt.style.use('fivethirtyeight')

# Creating the range for the x-axis values:
x_indexes = np.arange(len(ages_x)) 

# Creating a variable called x_indexes, that is an array of values which are gonna be the numbered version of the x-axis values
# and instead of using ages_x for the X data, we use x_indexes for the X data while plotting

#  Creating a width variable to make an offset for each X data for each plot. without this, the bars would still be stacked upon each other
width = 0.25

# Adding width of the bars equal to the width we define, just keep everything uniform
plt.bar(x_indexes - width, dev_y, width = width, color='#444444', label = "All devs")   # Offsetting to the left side by subtracting width

plt.bar(x_indexes,py_dev_y,  width = width,color="#008fd5", label ="Python")  # keeping in the middle (not giving any width)

plt.bar(x_indexes + width, js_dev_y,  width = width,color = '#e5ae38', label = "JavaScript") # Offsetting to the right side by adding width

plt.legend()

plt.xlabel("Ages")
plt.ylabel("Median salary in USD")
plt.title("Median salary in USD by Age")

plt.tight_layout()  # for the padding

plt.show()
```

Now we can see all the bars are side by side instead of stacking over each other.

**Notice that we no longer have the age ranges that we used to have, there is indexes now which we defined to do the offset part.**

We need to fix this..... using plt.xticks()

```python
plt.style.use('fivethirtyeight')

# Creating the range for the x-axis values:
x_indexes = np.arange(len(ages_x)) 

# Creating a variable called x_indexes, that is an array of values which are gonna be the numbered version of the x-axis values
# and instead of using ages_x for the X data, we use x_indexes for the X data while plotting

#  Creating a width variable to make an offset for each X data for each plot. without this, the bars would still be stacked upon each other
width = 0.25

# Adding width of the bars equal to the width we define, just keep everything uniform
plt.bar(x_indexes - width, dev_y, width = width, color='#444444', label = "All devs")   # Offsetting to the left side by subtracting width

plt.bar(x_indexes,py_dev_y,  width = width,color="#008fd5", label ="Python")  # keeping in the middle (not giving any width)

plt.bar(x_indexes + width, js_dev_y,  width = width,color = '#e5ae38', label = "JavaScript") # Offsetting to the right side by adding width

plt.legend()

plt.xticks(ticks=x_indexes, labels = ages_x)  # reintroducing the age ranges in the x-axis values 
# notice that labels that we use here is different from what we use for the plt.legend()

plt.xlabel("Ages")
plt.ylabel("Median salary in USD")
plt.title("Median salary in USD by Age")

plt.tight_layout()  # for the padding

plt.show()
```

Now we have the age ranges as our x-axis values

# Using Counter module to count the number of times language appeared in our csv file data:

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from collections import Counter

plt.style.use("fivethirtyeight")

# Loading csv file using csv module:
with open(r"C:\Users\JERRY\Jupyter_Python_2K24_25\Matplotlib Python\Data\data.csv") as csv_file:
    csv_reader = csv.DictReader(csv_file)   # reading file in a dictionary viewer so that we can call the key instead of the value

    language_counter = Counter()

    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
        
# # print(language_counter)
# print(language_counter.most_common(15))  # print most common items (inbuilt in Counter module)

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

# print(languages)
# print(popularity)  # Now we have two lists of languages and count

# plotting the data:
plt.bar(languages,popularity)

plt.title('Most Popular Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Number of People Who Use')

plt.tight_layout()

plt.show()
    
```

Notice that things are not clearly visible in the vertical bar chart...
