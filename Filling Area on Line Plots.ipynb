# Filling Area on Line Plots:
Fills are not only good for making our plot more professional but also it gives us the useful insights on how to use them.

for example, we can use conditional fills to show us what areas of our line plot fall above or below a certain threshold.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\JERRY\Jupyter_Python_2K24_25\Matplotlib Python\Data\data_fill betweens.csv")
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# filling the area below the py_devs line
plt.fill_between(ages, py_salaries, alpha= 0.25) 

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD')

plt.tight_layout()

plt.show()
```

Right now our fill is defaulting to zero which means it is filling between our y values to all the way to the bottom.

so we have the data for overall median salary for all developers who answered the survey.


```python
data = pd.read_csv(r"C:\Users\JERRY\Jupyter_Python_2K24_25\Matplotlib Python\Data\data_fill betweens.csv")
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

# filling the area below the py_devs line
plt.fill_between(ages, py_salaries, overall_median, alpha= 0.25) 

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD')

plt.tight_layout()

plt.show()
```

So now it showing us exactly where our python plot crossed that overall median salary. So when our python plot is below the overall median salary it fills up to that point and when our plot is above the overall median salary, it fills down to that point.

### We can also put some additional conditionals here, which only fills when certain condition is met.
where

```python
data = pd.read_csv(r"C:\Users\JERRY\Jupyter_Python_2K24_25\Matplotlib Python\Data\data_fill betweens.csv")
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

# filling the area below the py_devs line
plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries>overall_median), alpha= 0.25,
                interpolate = True)  # This will jsut make sure that certain x-intersects dont get clift and all the rigions fills properly

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD')

plt.tight_layout()

plt.show()
```

We can see here that we are only doing that fill where our python plot above the overall median salary. 

### Adding different color fills in the chart:

```python
data = pd.read_csv(r"C:\Users\JERRY\Jupyter_Python_2K24_25\Matplotlib Python\Data\data_fill betweens.csv")
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

# filling the area below the py_devs line
plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries > overall_median), color = 'yellow', alpha= 0.25,
                interpolate = True) 

# filling the area below the py_devs line
plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries <= overall_median), color = 'red', alpha= 0.25,
                interpolate = True)  # This will jsut make sure that certain x-intersects dont get clift and all the rigions fills properly

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD')

plt.tight_layout()

plt.show()
```

### Filling the area between the two Plots and Adding labels for the fill regions:
label

```python
data = pd.read_csv(r"C:\Users\JERRY\Jupyter_Python_2K24_25\Matplotlib Python\Data\data_fill betweens.csv")
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

# filling the area below the py_devs line
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries > dev_salaries), color = 'yellow', alpha= 0.25, label = 'Above_Avg',
                interpolate = True) 

# filling the area below the py_devs line
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries <= dev_salaries), color = 'red', alpha= 0.25, label = 'Below_Avg',
                interpolate = True)  # This will jsut make sure that certain x-intersects dont get clift and all the rigions fills properly

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD')

plt.tight_layout()

plt.show()
```
