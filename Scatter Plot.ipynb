# Scatter Plot:
Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to show how much one variable is affected by another. Each row in the data table is represented by a marker the position depends on its values in the columns set on the X and Y axes. A third variable can be set to correspond to the color or size of the markers, thus adding yet another dimension to the plot.

https://www.tutorialspoint.com/matplotlib/matplotlib_scatter_plot.htm

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x,y, s=100, c='green', marker='X', edgecolor='black', linewidth=1, alpha=0.75) 
# s-> size of the dots 
# c-> color of the dots
# markers -> replacing the dots with other markers
# alpha -> gives transperancy to the markers

plt.tight_layout() # add some automatic padding to the plot for better representation of the plot.

plt.show()
```

## Adding colormaps to the markers:
cmap

```python
plt.style.use('fivethirtyeight')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x,y, s=100, c= colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75) 


plt.tight_layout() # add some automatic padding to the plot for better representation of the plot.

plt.show()
```

Plotting the values of x and y and choosing the color of corresponding [Colors] value color. e.g. plotting 5 and 7 and choosing the color of 7. ligter dots are more likely to be closer to zero and darker one more likely to be closer to 10

## Adding a color bar for better understanding of the marker's colors:
cbar

```python
plt.style.use('fivethirtyeight')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x,y, s=100, c= colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75) 

cbar = plt.colorbar()
cbar.set_label('Satisfaction Level')

plt.tight_layout() # add some automatic padding to the plot for better representation of the plot.

plt.show()
```

## Defining different sizes of the markers:

```python
plt.style.use('fivethirtyeight')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

# Sizes of the markers corresponding to the datapoints in the x and y variables
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x,y, s=sizes, c= colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75) 

cbar = plt.colorbar()
cbar.set_label('Satisfaction Level')



plt.tight_layout() # add some automatic padding to the plot for better representation of the plot.

plt.show()
```

## Now Using CSV file data for Plotting:

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn-v0_8')

# Data of YouTube Video analysis:
data = pd.read_csv(r"C:\Users\JERRY\Jupyter_Python_2K24_25\Matplotlib Python\Data\data_scatter_plot.csv")
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
