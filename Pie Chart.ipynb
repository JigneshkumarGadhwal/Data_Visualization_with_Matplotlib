# Pie Chart:

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

```python
import matplotlib.pyplot as plt 
```

```python
plt.style.use('fivethirtyeight')

slices = [60,40,30,20]
# slices = [120,80]
labels = ['Sixty','Forty','extra_1','extra_2']  # the order would be correspond to the order we define in the slices
# NOTE: our values do not need to be sum up to 100 %, they can be anything and pie chart would calculate the proportion automatically.

# adding colors to the pie portions:
colors = ['blue','red','yellow','green'] # we can also define hex colors.

plt.pie(slices,labels = labels, colors =colors, wedgeprops={"edgecolor": "black"})  #wedgeprops ->edgecolor adds borders in the pie chart

plt.title('My Amazing Pie Chart')
plt.tight_layout()
plt.show()
```

**IMPORTANT NOTE:** our data slices = [60,40] represented as the values and in pie chart is represented as percentage of the data.
for example if we use [120,80] , we would still be getting the same percentage in the pie chart.

### Emphasizing the portion in pie chart:
explode

```python
plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917] # Users
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0] # list of floats which represents how much we want to emphasize the portion. [0 is default, other thatn that will define the fraction of the radius that we want ot emphasize]
                        # ephasizing Python which is 4th position, 0.1 -> 10 % of the radius is exploded in the pie chart 

plt.pie(slices,labels = labels, explode = explode,
        wedgeprops={"edgecolor": "black"})  #wedgeprops ->edgecolor adds borders in the pie chart

plt.title('My Amazing Pie Chart')
plt.tight_layout()
plt.show()
```

### Adding shadows and starting angle in the Pie Chart:
shadow, startangle

```python
plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917] # Users
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0] # list of floats which represents how much we want to emphasize the portion. [0 is default, other thatn that will define the fraction of the radius that we want ot emphasize]
                        # ephasizing Python which is 4th position, 0.1 -> 10 % of the radius is exploded in the pie chart 

plt.pie(slices,labels = labels, 
        explode = explode, 
        shadow = True,  # adding shadow in the pie chart
        startangle=90, # notice the first value for JS starts at 90 degree angle and accordingly...
        wedgeprops={"edgecolor": "black"})  #wedgeprops ->edgecolor adds borders in the pie chart

plt.title('My Amazing Pie Chart')
plt.tight_layout()
plt.show()
```

### Adding percentages for the portions in the pie chart:
autopct

```python
plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917] # Users
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0] # list of floats which represents how much we want to emphasize the portion. [0 is default, other thatn that will define the fraction of the radius that we want ot emphasize]
                        # ephasizing Python which is 4th position, 0.1 -> 10 % of the radius is exploded in the pie chart 

plt.pie(slices,labels = labels, 
        explode = explode, 
        shadow = True,  # adding shadow in the pie chart
        startangle=90, # notice the first value for JS starts at 90 degree angle and accordingly...
        wedgeprops={"edgecolor": "black"},  #wedgeprops ->edgecolor adds borders in the pie chart
        autopct = '%1.1f%%')

plt.title('My Amazing Pie Chart')
plt.tight_layout()
plt.show()
```
